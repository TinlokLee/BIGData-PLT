####  美国大选数据分析

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
%matplotlib inline
```

```python
ele = pd.read_csv('./usa_election.csv')
ele.head()
```

```python
months = {'JAN' : 1, 'FEB' : 2, 'MAR' : 3, 'APR' : 4, 'MAY' : 5, 'JUN' : 6,
          'JUL' : 7, 'AUG' : 8, 'SEP' : 9, 'OCT': 10, 'NOV': 11, 'DEC' : 12}
of_interest = ['Obama, Barack', 'Romney, Mitt', 'Santorum, Rick', 
               'Paul, Ron', 'Gingrich, Newt']
parties = {
  'Bachmann, Michelle': 'Republican',
  'Romney, Mitt': 'Republican',
  'Obama, Barack': 'Democrat',
  "Roemer, Charles E. 'Buddy' III": 'Reform',
  'Pawlenty, Timothy': 'Republican',
  'Johnson, Gary Earl': 'Libertarian',
  'Paul, Ron': 'Republican',
  'Santorum, Rick': 'Republican',
  'Cain, Herman': 'Republican',
  'Gingrich, Newt': 'Republican',
  'McCotter, Thaddeus G': 'Republican',
  'Huntsman, Jon': 'Republican',
  'Perry, Rick': 'Republican'           
 }
```

```python
#  表格添加党派信息
#  2.映射：用map将候选人映射到对应党派，新列
```

```python
ele["party"]=ele['cand_nm'].map(parties)
ele.head()
```

```python
ele.party.unique()
```

```python
#  查看各党派政治捐献总金额
#  以党派分组，对政治捐献聚合-求和
```

```python
ele.groupby(["party"])[["contb_receipt_amt"]].sum()
```

```python
#   查看具体每一天，各个党派收到的政治捐献
#   以党派和日期作为分组，对政治捐献进行聚合
```

```python
ele.groupby(["contb_receipt_dt","party"])[['contb_receipt_amt']].sum()
```

```python
#  对日期排序
#ele.dtypes  日期是字符串类型，转为日期类型
pd.to_datetime(['01-APR-11'])
#   耗时 pd.to_datetime(ele['contb_receipt_dt'])  
```

```python
#  将日期格式化,标准化
#  replace不接函数，map()后映射
def conver_date(date):
    day,moth,year=date.split('-')
    m = moths[moth]
    return "20"+ str(year) + '-'+ str(m)+'-'+ str(day)

```

```python
ele['contb_receipt_dt']=ele['contb_receipt_dt'].map(conver_date)
```

```python
ele.head()
```

```python
ele['contb_receipt_dt']=pd.to_datetime['contb_receipt_dt']
```

```python
ele.dtypes
```

```python
ele.sort_values('contb_receipt_dt',inplace=True)
```

```python
#  排序后每一天每个党派的政治捐献
```

```python
ele.groupby(['contb_receipt_dt','party'])[['contb_receipt_amt']].sum()
```
